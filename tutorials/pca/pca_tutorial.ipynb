```python
import numpy as np
import matplotlib.pyplot as plt
```

PCAのライブラリのインポート

```python
from sklearn.decomposition import PCA
```

```python
seed = 10
np.random.seed(seed)
mean = np.zeros(2)
cov = np.array([[1,0.5],[0.5,1]])
X = np.random.multivariate_normal(mean, cov, 500)
X = X - X.mean(axis=0)
```

```python
pca = PCA(n_components=2)
pca.fit(X)
```

```python
W = pca.components_
```

```python
fig = plt.figure(0)
ax = fig.add_subplot('111', aspect='equal')
ax.scatter(X[:,0],X[:,1],marker='o',s=10, c=X[:,0])
ax.quiver(np.zeros(2),np.zeros(2),W[:,0],W[:,1])
```

```python
plt.show()
```

```python
Xd = pca.transform(X)
np.allclose(Xd,X@W.T)
```

```python
fig = plt.figure(1)
ax = fig.add_subplot('111', aspect='equal')
ax.scatter(Xd[:,0],Xd[:,1],marker='o',s=10,c=X[:,0])
```

```python
plt.show()
```
