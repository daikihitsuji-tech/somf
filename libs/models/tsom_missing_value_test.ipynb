```python
import numpy as np
import pandas as pd

# 10列 x 10行の行列
raw_data = np.random.rand(10, 10)

# DataFrameに変換
dataset = pd.DataFrame(raw_data)
dataset
```

```python
def remove_random_cell(dataframe):
    # 欠損させるindexを選ぶ
    target_index = np.random.choice(dataframe.index)

    # 欠損させるcolumnを選ぶ
    target_column = np.random.choice(dataframe.columns)

    # 欠損させる
    dataframe.loc[target_index, target_column] = None
    
# 100回実行
for _ in range(100):
    remove_random_cell(dataset)
    
dataset
```

```python
datasets = dataset.values
# このdatasets が欠損値ある場合の入力
datasets
```

```python
# nanならTrue
np.isnan(datasets)
```

```python
# True == 1, False == 0
a = np.where(np.isnan(datasets) == 1, 0, 1)
a
```

```python
# 欠損値判定 True:欠損値 があるか判定
np.any(np.isnan(datasets) == 1)
```

```python
datasets[np.isnan(datasets)] = 0 
datasets * a
```
